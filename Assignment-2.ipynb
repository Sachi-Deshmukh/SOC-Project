```python
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2):
    return (x1 - x2) ** 4 + 8 * x1 * x2 - x1 + x2 + 3

# Create a grid of x1 and x2 values
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)

# Calculate the values of f(x1, x2) for each combination of x1 and x2
Z = f(X1, X2)

# Generate the contour plot
plt.contour(X1, X2, Z, levels=20, cmap='jet')
plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot of f(x1, x2)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points
years_experience = [1, 3, 5]
salary = [300, 480, 570]

# Given parameters
w = 200
b = 100

# Scatter plot
plt.scatter(years_experience, salary)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')

# Plot the line with given parameters
x = [0, max(years_experience)]
y = [w * xi + b for xi in x]
plt.plot(x, y, color='red', label='Line: w = 200, b = 100')

# Display legend
plt.legend()

# Display the plot
plt.show()
```

```python
def compute_cost(years_experience, salary, w, b):
    total_cost = 0
    m = len(years_experience)  # Number of data points

    for i in range(m):
        x = years_experience[i]
        y = salary[i]
        predicted_salary = w * x + b
        cost = (predicted_salary - y) ** 2
        total_cost += cost

    total_cost /= (2 * m)  # Average the total cost

    return total_cost
years_experience = [1, 3, 5]
salary = [300, 480, 570]
w = 200
b = 100

cost = compute_cost(years_experience, salary, w, b)
print("Total cost:", cost)
```

```python
def compute_gradient(years_experience, salary, w, b):
    dw = 0
    db = 0
    m = len(years_experience)  # Number of data points

    for i in range(m):
        x = years_experience[i]
        y = salary[i]
        predicted_salary = w * x + b
        dw += (predicted_salary - y) * x
        db += (predicted_salary - y)

    dw /= m
    db /= m

    return dw, db
years_experience = [1, 3, 5]
salary = [300, 480, 570]
w = 200
b = 100

dw, db = compute_gradient(years_experience, salary, w, b)
print("Gradient with respect to w:", dw)
print("Gradient with respect to b:", db)
```

```python
def compute_gradient_descent(x_train, y_train, w_init, b_init, alpha, num_iters):
    w = w_init
    b = b_init

    for i in range(num_iters):
        # Compute cost and gradients
        cost = compute_cost(x_train, y_train, w, b)
        dw, db = compute_gradient(x_train, y_train, w, b)

        # Update parameters using gradient descent
        w -= alpha * dw
        b -= alpha * db

        # Print cost every 1000 iterations
        if i % 1000 == 0:
            print(f"Iteration {i}: Cost = {cost}")

    return w, b
x_train = [1, 3, 5]
y_train = [300, 480, 570]
w_init = 0
b_init = 0
alpha = 0.01
num_iters = 10000

final_w, final_b = compute_gradient_descent(x_train, y_train, w_init, b_init, alpha, num_iters)
print("Final w:", final_w)
print("Final b:", final_b)
```
